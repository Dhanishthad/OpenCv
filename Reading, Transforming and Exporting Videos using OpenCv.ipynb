```python
#establish capture
cap = cv2.VideoCapture(os.path.join("C:/Users/Dhanishtha/OneDrive/Documents/Placement Preparation/Data Science/OpenCv/Data/Videos/Teacher Teaching the Students Inside a Classroom - Free Stock Videos.mp4"))

#properties
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
fps = cap.get(cv2.CAP_PROP_FPS)

#setup video writer
video_writer = cv2.VideoWriter(os.path.join("C:/Users/Dhanishtha/OneDrive/Documents/Placement Preparation/Data Science/OpenCv/Data/Output/output.avi"),cv2.VideoWriter_fourcc('P','I','M','1'), fps, (width,height), isColor=False)

#cv2.VideoWriter_fourcc('P','I','M','1'): combines these 4 char to come up with a codec which we are going to use 
#A FOURCC code is a 32-bit unsigned integer that is created by concatenating four ASCII characters. 
#For example, the FOURCC code for YUY2 video is 'YUY2'. 

#loop through each frame
for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    #Read frame
    ret, frame =cap.read()

    #gray transform
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    #show image 
    #cv2.imshow('video player', frame)
    cv2.imshow('video player', gray)

    #write out frame
    video_writer.write(gray)
    

    #breaking out of the loop
    #waitKey(10) checks whether we are hitting the keyboard key after some time interval
    #0xff gets what key we hit 'q' on keyboard
    #hitting 'q' shuts down the frame
    if cv2.waitKey(10) & 0xff == ord('q'):
        break

#close down everything
cap.release()
cv2.destroyAllWindows()

#release/close video writer
video_writer.release()
```
