```python
# Show image
plt.imshow(img)
#the flare here is blue whereas original flare is red in color
#this is because matplotlib expects the image to be in rgb format
#opencv reads it as bgr
#solution= reshape/recolor the color channels
```

```python
recolor = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(recolor)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
```
